```python
df=pd.read_csv('products.csv')
df.head()
```

```python
df.shape
```

```python
data=list(df["Products"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
df=apriori(df,min_support=0.02,use_colnames=True)
df
```

```python
df_ar=association_rules(df,metric="confidence", min_threshold=0.3)
df_ar
```

```python
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

```python
df_ar=association_rules(df,metric="conviction", min_threshold=1.0)
df_ar
```
