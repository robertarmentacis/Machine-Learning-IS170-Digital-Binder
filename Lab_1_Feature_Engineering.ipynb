```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the CSV (salary dataset)

df = pd.read_csv("Salary_Dataset_DataScienceLovers.csv")
df

# Deal with missing / null values
# 1) Drop missing values (example shown in PDF)

df1 = df.dropna()   # df1 is the "revised" dataset with no missing values :contentReference[oaicite:6]{index=6}


# 2) Replace the missing data with another value (fillna)
#    (PDF shows fillna used to recode missing values) :contentReference[oaicite:7]{index=7}

df = df.fillna(0)

# 3) Replace the zero values with the mean value (PDF maneuver) :contentReference[oaicite:8]{index=8}
#    We'll do it for Salary if any salaries are 0.
salary_mean = df.loc[df["Salary"] != 0, "Salary"].mean()
df["Salary"] = df["Salary"].replace(0, salary_mean)


# Convert String to Numeric Values (LabelEncoder)
# create temporary "label" column
# drop original categorical column
# save label column back under original name (PDF pattern) :contentReference[oaicite:9]{index=9}
le = LabelEncoder()

# Example target-like categorical: Employment Status
df["label"] = le.fit_transform(df["Employment Status"])

df = df.drop("Employment Status", axis=1)      # drop original
df["Employment Status"] = df["label"]          # save label as original column name
df = df.drop("label", axis=1)                  # remove temp column


# - standardization => mean 0, std 1
mm = MinMaxScaler()
ss = StandardScaler()

df["Salary_norm"] = mm.fit_transform(df[["Salary"]])
df["Salary_std"]  = ss.fit_transform(df[["Salary"]])

# One-Hot Encoding (simple version using get_dummies)
df = pd.get_dummies(df, columns=["Location", "Job Roles"], drop_first=False)


# Column Manipulation
# - In supervised learning, features first, target last (PDF) :contentReference[oaicite:12]{index=12}
# We'll treat Salary as the target and move it to the last column.
cols = [c for c in df.columns if c != "Salary"]

# Rename the column headers (PDF example) :contentReference[oaicite:13]{index=13}
df.columns = df.columns.str.lower().str.replace(" ", "_")

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

#Load the CSV file into a DataFrame
df = pd.read_csv("defects_data.csv")
df

#Clean text columns by removing leading/trailing spaces
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip()

#Convert defect_date column to datetime format
if "defect_date" in df.columns:
    df["defect_date"] = pd.to_datetime(df["defect_date"], errors="coerce")

#Fill missing values
#numeric columns -> median
#categorical columns -> "Unknown"
for col in df.select_dtypes(include=["number"]).columns:
    df[col] = df[col].fillna(df[col].median())

for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].replace(["nan", "", "None"], np.nan).fillna("Unknown")

#Replace repair_cost = 0 with the average non-zero cost
if "repair_cost" in df.columns:
    avg = df.loc[df["repair_cost"] != 0, "repair_cost"].mean()
    df["repair_cost"] = df["repair_cost"].replace(0, avg)

# Convert severity text labels into numbers

if "severity" in df.columns:
    le = LabelEncoder()
    df["severity"] = le.fit_transform(df["severity"])

#One-hot encode selected categorical columns as we added those into the dataset

cols = ["defect_type", "inspection_method"]
cols = [c for c in cols if c in df.columns]

#Changing defect_date to defected_date
df.rename(columns={'defect_date': 'defected_date'}, inplace=True)


```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Load data
df = pd.read_csv("Students Social Media Addiction.csv")
df

# Missing values (PDF style)
df1 = df.dropna()      # example: dropping missing rows
df = df.fillna(0)      # example: filling missing values with 0

# Replace 0 with mean (pick one numeric column)
num_col = df.select_dtypes(include=["int64", "float64"]).columns[0]
df[num_col] = df[num_col].replace(0, df[num_col].mean())

# Label encode (pick one categorical column)
cat_col = df.select_dtypes(include="object").columns[0]
le = LabelEncoder()
df["label"] = le.fit_transform(df[cat_col].astype(str))
df = df.drop(cat_col, axis=1)
df[cat_col] = df["label"]
df = df.drop("label", axis=1)

# Normalize + standardize (same numeric column)
mm = MinMaxScaler()
ss = StandardScaler()
df[num_col + "_norm"] = mm.fit_transform(df[[num_col]])
df[num_col + "_std"]  = ss.fit_transform(df[[num_col]])

# One-hot encode remaining categorical columns
df = pd.get_dummies(df)

# Rename headers (PDF example)
df.columns = df.columns.str.lower().str.replace(" ", "_")

```
