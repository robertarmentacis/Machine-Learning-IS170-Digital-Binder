```python
import pandas as pd                      # Used for working with tables/dataframes
import numpy as np                       # Used for numerical operations
import matplotlib.pyplot as plt          # Used for plotting graphs if needed

from sklearn.tree import DecisionTreeClassifier   # Decision Tree model
from sklearn.ensemble import RandomForestClassifier  # Random Forest model
from sklearn.model_selection import train_test_split # Splits data into training/testing sets
from sklearn import metrics              # Used for model evaluation
from sklearn.preprocessing import LabelEncoder     # Converts text labels into numbers
```

```python
# Load the dataset
# Reads the uploaded CSV file into a pandas dataframe

df = pd.read_csv("data.csv")

# Show a preview of the dataset
print("Dataset preview:")
df.head()
```

```python
# Data cleaning
# Remove columns that are not useful for prediction

df.drop("id", axis=1, inplace=True)           # Remove ID column
df.drop("Unnamed: 32", axis=1, inplace=True)  # Remove empty/unnecessary column
```

```python
# Encode the target variable
# The diagnosis column contains text labels like:
# M = malignant
# B = benign
# Machine learning models need numbers, so we convert them.

le = LabelEncoder()

label = le.fit_transform(df["diagnosis"])   # Convert diagnosis text into numeric values
df["diagnosis"] = label                     # Replace original text column with encoded values

# Display results
print(label)
```

```python
#split dataset in features and target variable
x_data=df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean',]]
y_data=df['diagnosis']

```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
```

```python
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train,y_train)

# Calculate the accuracy and assign it to dt_accuracy
# Train a single Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Calculate the accuracy and assign it to dt_accuracy
dt_accuracy = dt_model.score(X_test, y_test)

print("Decision Tree Test Accuracy:", f"{dt_accuracy * 100:.2f}%")
```

```python
# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split

rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)


# Train a Random Forest with random feature selection
rf_model = RandomForestClassifier(n_estimators=5, max_features=2, random_state=42)  # max_features=2 means only 2 features are considered at each split


rf_model.fit(X_train, y_train)


# Calculate the accuracy and assign it to rf_accuracy
rf_accuracy = rf_model.score(X_test, y_test)

# Now you can use rf_accuracy in the print statements below
print("RF Model Accuracy:", f"{rf_accuracy * 100:.2f}", "%")
```

```python
# Prepare a section for a new data entry and its prediction of an outcome

def predict_cancer():
    while True:
        try:
            radius_mean = float(input("Enter radius_mean: "))
            texture_mean = float(input("Enter texture_mean: "))
            perimeter_mean = float(input("Enter perimeter_mean: "))
            area_mean = float(input("Enter area_mean: "))
            smoothness_mean = float(input("Enter smoothness_mean: "))
            compactness_mean = float(input("Enter compactness_mean: "))

            # Input validation
            if any(val < 0 for val in [
                radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean,
                compactness_mean
            ]):
                print("Error: Please enter non-negative values for all features.")
                continue

            new_patient_data = pd.DataFrame([[
                radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean,
                compactness_mean
            ]], columns=[
                'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
                'compactness_mean'
            ])

            prediction = rf_model.predict(new_patient_data)[0]

            if prediction == 1:
                print("Prediction: The tumor is likely MALIGNANT.")
            else:
                print("Prediction: The tumor is likely BENIGN.")

            another_patient = input("Enter data for another patient? (yes/no): ")
            if another_patient.lower() != 'yes':
                break

        except ValueError:
            print("Invalid input. Please enter valid numerical values.")

predict_cancer()
```

**Random Forest dataset Report**

This project developed a machine learning model to help classify tumors as either benign (non-cancerous) or malignant (cancerous) using diagnostic measurements from a breast cancer dataset. The dataset was first cleaned by removing unnecessary columns such as identification numbers and empty variables, and the diagnosis labels were converted from text values (“M” and “B”) into numerical values so the algorithms could process them. Six key medical features were selected for prediction: radius mean, texture mean, perimeter mean, area mean, smoothness mean, and compactness mean, which describe the tumor’s size, shape, and surface characteristics. The data was split so that 70% was used to train the models and 30% was used to test their performance, ensuring the models were evaluated on unseen data. Two classification models were developed: a Decision Tree and a Random Forest. The Decision Tree achieved about 88% accuracy, while the Random Forest produced a similar level of accuracy and additional evaluation metrics including precision, recall, and F1-score, which help measure how effectively the model identifies malignant tumors while minimizing incorrect predictions. The system also includes a feature that allows new patient measurements to be entered and evaluated in real time to generate a prediction. A visualization of the Decision Tree shows how the model makes decisions step-by-step; for example, a node might evaluate whether perimeter_mean ≤ 114.45, meaning the model checks if the tumor’s average perimeter is below a certain threshold to decide which branch to follow. Each node also displays values such as gini (which measures how well the data is separated), samples (the number of cases reaching that node), and value (the distribution of benign versus malignant cases), helping explain how the model arrives at its final classification.
