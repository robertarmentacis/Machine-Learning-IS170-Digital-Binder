```python
# RSB Kernal Multiclass
# Import necessary libraries
import pandas as pd              # Data handling (tables, CSV files)
import numpy as np               # Numerical operations

from sklearn import datasets     # Built-in datasets (not used later here)
from sklearn.model_selection import train_test_split  # Split data into train/test
from sklearn.svm import SVC      # Support Vector Machine model
from sklearn.metrics import accuracy_score, confusion_matrix  # Evaluation metrics
import matplotlib.pyplot as plt  # Plotting graphs
import seaborn as sns            # Better visualization (heatmaps, etc.)
from sklearn.preprocessing import LabelEncoder  # Convert text to numbers

# Load the digits dataset (NOT used later, just loaded)
digits = datasets.load_digits()

```

```python
# Load dataset from CSV file
# Upload the dataset file from your computer
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("test.csv")

# Preview first few rows
df.head()

# Drop unnecessary columns if they exist
df = df.drop(columns=["unnamed: 0", 'id'], errors='ignore')

# Preview again after dropping columns
df.head()
```

```python
# Drop column that has missing values or not needed
df.drop("Arrival Delay in Minutes", axis=1, inplace=True)


```

```python
# Encode categorical (text) columns into numbers
label_encoders = {}  # Store encoders for later (important for decoding)

# Find all columns with text data
categorical_columns = df.select_dtypes(include=["object"]).columns

# Loop through each categorical column and encode it
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # Convert text → numbers
    label_encoders[col] = le            # Save encoder


# Define features (X) and target (y)
# Target variable = satisfaction
X = df.drop("satisfaction", axis=1)  # All columns except target
y = df["satisfaction"]               # Target column

```

```python
# Split data into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


# Create SVM model with RBF kernel
# RBF = handles non-linear relationships
svm_classifier = SVC(kernel='rbf', gamma='scale', C=1.0)

# Train the model on training data
print("Training the SVM model...")
svm_classifier.fit(X_train, y_train)
print("Training complete!")

```

```python
# Make predictions using test data
y_pred = svm_classifier.predict(X_test)

# Calculate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy * 100:.2f}%")

```

```python
# Create confusion matrix (shows correct vs incorrect predictions)
cm = confusion_matrix(y_test, y_pred)

# Get original class labels (before encoding)
target_names = label_encoders["satisfaction"].classes_

# Plot confusion matrix as heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,                  # Show numbers inside boxes
    fmt='d',                     # Integer format
    cmap='Blues',                # Color scheme
    xticklabels=target_names,    # Predicted labels
    yticklabels=target_names     # Actual labels
)
plt.xlabel('Predicted Satisfaction')
plt.ylabel('Actual Satisfaction')
plt.title('Confusion Matrix: Airline Satisfaction Classification')
plt.show()


```

```python
# Show some sample predictions
print("\nSample Predictions:")

# Copy test data and add actual + predicted values
sample_results = X_test.copy()
sample_results["Actual"] = y_test
sample_results["Predicted"] = y_pred

# Convert encoded numbers back to original labels (more readable)
sample_results["Actual"] = label_encoders["satisfaction"].inverse_transform(sample_results["Actual"])
sample_results["Predicted"] = label_encoders["satisfaction"].inverse_transform(sample_results["Predicted"])

# Display first 10 results
print(sample_results[["Actual", "Predicted"]].head(10))


# Print summary statistics
print(f"\nTotal test samples: {len(y_test)}")
print(f"Correctly classified: {(y_test == y_pred).sum()}")
print(f"Incorrectly classified: {(y_test != y_pred).sum()}")
```
