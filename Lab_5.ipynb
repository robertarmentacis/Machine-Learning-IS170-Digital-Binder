```python
# Import the necessary libraries for data and graphing
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('ai_worker_burnout_attrition_2026.csv')

# Show the first few rows to check the data
df.head()
```

```python
# Select the independent variable (X) and dependent variable (y)
# We are using 'ai_replaces_my_tasks_pct' to predict 'burnout_score'
x = df[['ai_replaces_my_tasks_pct']].values
y = df['burnout_score'].values

# Split the data into training (75%) and testing (25%) sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# Import and build the standard Linear Regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Create the model
model = LinearRegression()

# Train the model using the training data
model.fit(x_train, y_train)

# Make predictions on the test data
y_pred = model.predict(x_test)

# Check how accurate the model is
accuracy = r2_score(y_test, y_pred)
print(f"Standard Model Accuracy (R-Squared): {accuracy}")
```

```python
# Optimize for the degree value  to see if a curved line fits better
from sklearn.preprocessing import PolynomialFeatures

# We use degree=3 for this test to allow the model to learn nonlinear relationships
poly = PolynomialFeatures(degree=3, include_bias=False)
x_poly = poly.fit_transform(x)

poly_model = LinearRegression()
poly_model.fit(x_poly, y)

# Create a visual plot for stakeholders
x_range = nm.linspace(x.min(), x.max(), 100).reshape(-1, 1)
y_range_pred = poly_model.predict(poly.transform(x_range))

# Plot the graph

#Image of Linear Regression

plt.scatter(x, y, color='blue', alpha=0.3, label='Worker Data')
plt.plot(x_range, y_range_pred, color='red', label='Polynomial Trend Line')
plt.xlabel('AI Task Replacement (%)')
plt.ylabel('Burnout Score')
plt.title('AI Automation vs. Employee Burnout')
plt.legend()
plt.show()
```

This analysis used a linear regression model to evaluate a worker's anticipated burnout score based on the percentage of their daily tasks replaced by AI tools. The dataset was prepared by isolating the task replacement and burnout score variables, and splitting the data into training and testing groups so the model could be trained and then evaluated on unseen data. Based on the results, the model achieved an R-Squared score of about 35.9% and a Mean Squared Error of 78.6. This tells us the model successfully identifies a clear trend, showing that as AI task replacement increases, burnout scores predictably rise even though it does not account for all the different factors that cause burnout. In practical terms, this can be valuable in an organizational setting to anticipate workforce fatigue when introducing new AI automation, where catching early burnout is crucial for employee retention. Overall, the analysis shows that task replacement measurements do contain useful patterns for predicting burnout risk, although the model could likely be improved further by including additional workplace variables and upgrading to a multivariate approach.

```python
#Import Libraries
# numpy for numeric calculations
import numpy as np

# pandas for reading and working with the dataset
import pandas as pd

# logistic regression model
from sklearn.linear_model import LogisticRegression

# used to split data into training and testing sets
from sklearn.model_selection import train_test_split

# used to scale the feature values
from sklearn.preprocessing import StandardScaler

# tools used to evaluate model performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Read the coffee dataset CSV file
df = pd.read_csv("df_arabica_clean.csv")

# Print basic dataset information
print("=" * 60)
print("COFFEE DATASET OVERVIEW")
print("=" * 60)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

# Show the first 5 rows to understand the data
print("\nFirst 5 rows:")
print(df.head())

# Remove extra column if it exists
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# These columns will be used as input variables
features = [
    "Aroma",
    "Flavor",
    "Aftertaste",
    "Acidity",
    "Body",
    "Balance",
    "Uniformity",
    "Clean Cup",
    "Sweetness",
    "Overall",
    "Moisture Percentage",
    "Defects",
    "Category One Defects",
    "Quakers",
    "Category Two Defects"
]

# X = input features for the model
X = df[features].copy()


```

```python
# Convert Total Cup Points score into 3 categories
# 0 = Low quality
# 1 = Medium quality
# 2 = High quality
def make_quality_class(score):
    if score < 82.5:
        return 0
    elif score < 84.5:
        return 1
    else:
        return 2

# y = the value we want to predict
y = df["Total Cup Points"].apply(make_quality_class)

# Class names for easier reading in output
class_names = ["Low", "Medium", "High"]

# Show how many samples are in each class
print("\nClass distribution:")
for i, name in enumerate(class_names):
    count = np.sum(y == i)
    print(f"{name}: {count} samples ({count/len(y)*100:.1f}%)")
```

```python
# Split the dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

print("\n" + "=" * 60)
print("TRAIN/TEST SPLIT")
print("=" * 60)
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size:  {X_test.shape[0]} samples")
```

```python
# StandardScaler makes the feature values comparable
scaler = StandardScaler()

# Fit scaler on training data
X_train_scaled = scaler.fit_transform(X_train)

# Apply the same scaling to test data
X_test_scaled = scaler.transform(X_test)

print(f"\nBefore scaling - Aroma mean: {X_train['Aroma'].mean():.2f}")
print(f"After scaling  - Aroma mean: {X_train_scaled[:, 0].mean():.4f}")


```

```python
# Create the model
model = LogisticRegression(
    solver="lbfgs",
    max_iter=1000,
    C=1.0,
    random_state=42
)

# Train the model using the training data
model.fit(X_train_scaled, y_train)

print("\n" + "=" * 60)
print("MODEL TRAINING COMPLETE")
print("=" * 60)
print(f"Number of iterations used: {model.n_iter_[0]}")
print(f"Coefficients shape: {model.coef_.shape}")


```

```python
# Predict class labels for the test data
y_pred = model.predict(X_test_scaled)

# Predict probabilities for each class
y_proba = model.predict_proba(X_test_scaled)

print("\n" + "=" * 60)
print("PREDICTIONS (First 5 Test Samples)")
print("=" * 60)

y_test_array = np.array(y_test)

# Show first 5 predictions
for i in range(min(5, len(y_test_array))):
    proba_dict = {
        class_names[j]: f"{y_proba[i][j]:.3f}"
        for j in range(len(class_names))
    }

    print(
        f"Sample {i+1}: Actual={class_names[y_test_array[i]]}, "
        f"Predicted={class_names[y_pred[i]]}, "
        f"Probabilities={proba_dict}"
    )
```

```python
# Calculate overall accuracy
accuracy = accuracy_score(y_test, y_pred)

print("\n" + "=" * 60)
print("MODEL EVALUATION")
print("=" * 60)
print(f"Overall Accuracy: {accuracy:.4f} ({accuracy*100:.1f}%)")

# Confusion matrix shows correct vs incorrect predictions
cm = confusion_matrix(y_test, y_pred)

print("\nConfusion Matrix:")
print("                Predicted")
print("             Low  Medium  High")
for i, name in enumerate(class_names):
    print(f"Actual {name:>6}: {cm[i]}")

# Classification report shows precision, recall, and f1-score
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
print("=" * 60)
print("FEATURE IMPORTANCE (Top 3 per class)")
print("=" * 60)

# Put model coefficients into a table
coef_df = pd.DataFrame(
    model.coef_,
    columns=features,
    index=class_names
)

# Find the top 3 features for each class
for class_name in class_names:
    top_features = coef_df.loc[class_name].abs().nlargest(3)

    print(f"\n{class_name}:")
    for feat, val in top_features.items():
        actual_val = coef_df.loc[class_name, feat]
        direction = "↑ increases" if actual_val > 0 else "↓ decreases"

        print(f"  {feat}: {actual_val:.3f} ({direction} probability)")


print("\n" + "=" * 60)
print("DONE! Coffee classification complete.")
print("=" * 60)
```
