```python
# call the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Get all the modules for the code

#Load the data
df=pd.read_csv("mydata.csv")
df.head()

#Scatter plot with the red dots size 18
# from the data, x column values and y columns
plt.scatter(df['x'],df['y'],c='red', s=18)


#K is the number of cluster we want
k=2

#Matrix containing all the points
X=np.array(list(zip(df['x'],df['y'])))

model=KMeans(n_clusters=k)

# To train the model using the fit() function
kmeans=model.fit(X)

# The predict function() and assign the labels to all of the points
labels=model.predict(X)

# To get the centroids, and use the cluster centers property
centroids=model.cluster_centers_

#This prints the clusters, label and centriods
print(labels)
print(centroids)

#Ints and centriods on a scatter plot
# map the labels to colors
c=['blue','red','yellow','green']
colors=[c[i] for i in labels]

#using the model that you have just trained,
#Just  can use it to predict what cluster for any point
# Making predictions
plt.scatter(df['x'],df['y'],c=colors, s=18)
plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=100, c='black')


cluster=model.predict([[3,4]])[0]
print(c[cluster])

cluster=model.predict([[7,5]])[0]
print(c[cluster])

cluster=model.predict([[7,9]])[0]
print(c[cluster])

# Obtaining the silhouette coefficient (score)

from sklearn.metrics import silhouette_score

model=KMeans(n_clusters=2)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=2):  {silhouette_score(X,label)}')
```

```python
# call the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# load the data
df=pd.read_csv("bermuda_shorts.csv")
df.head()

# info on the data
df.info()

# to know all the columns and rows
print(df.shape)

#To find how many na there is in the dataset
df.isnull().sum()

#remove the missing values with NaNs
df=df.dropna(subset=['bmxleg','bmxwaist'])
print(df.shape)

#scatter plot
plt.scatter(df['bmxleg'],df['bmxwaist'], c='red', s=2)
plt.xlabel("upper leg length (cm)")
plt.ylabel("waist circumference (cm)")

#For now, we'll try k value 2
k=2
X=np.array(list(zip(df['bmxleg'], df['bmxwaist'])))
model=KMeans(n_clusters=k)
kmeans=model.fit(X)
labels=model.predict(X)
centroids=model.cluster_centers_

print(labels)
print(centroids)


#Plot the data and centroids on a scatter plot
#map the labels to colors

c=['m','g','b','r','y','c']
colors=[c[i] for i in labels]

plt.scatter(df['bmxleg'], df['bmxwaist'], c=colors, s=2)
plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=100, c='black')

#Using the model that you have just trained,
#you can use it to predict what cluster for any data pt

cluster=model.predict([[35,123]])[0]
print(c[cluster])


#To determine the silhouette coefficient/score

from sklearn.metrics import silhouette_score

model=KMeans(n_clusters=2)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=2): {silhouette_score(X,label)}')


model=KMeans(n_clusters=3)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=3): {silhouette_score(X,label)}')


model=KMeans(n_clusters=4)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=4): {silhouette_score(X,label)}')


k=4
X=np.array(list(zip(df['bmxleg'],df['bmxwaist'])))

kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(X)
labels=kmeans.predict(X)
centroids=kmeans.cluster_centers_

#map the labels to colors
c=['b','r','y','g','c','m']
colors=[c[i] for i in labels]

plt.scatter(df['bmxleg'],df['bmxwaist'],c=colors,s=2)
plt.scatter(centroids[:,0],centroids[:,1],marker='*',s=100,c='black')

print(centroids)

```

```python
#Robert Armenta
#https://www.kaggle.com/datasets/wajahat1064/healthcare-appointment-dataset

#get all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
# Get all the modules for the code

#Load the data
df=pd.read_csv("healthcare_noshows.csv")
df.head()
df
# creating a instance of LabelEncoder
le = LabelEncoder()

df.info()

#using the fit.tranform function to fit label
label = le.fit_transform(df['Scholarship'])
#Then print label
label

#Removing the Scholarship from df
df.drop('Scholarship', axis=1, inplace=True)

#appending the array to the dataframe with Scholarship
df["Scholarships"]= label

# printing the dataframe
df

label=le.fit_transform(df["SMS_received"])

# using the fit.tranform function to labe
label=le.fit_transform(df["SMS_received"])

#removing the column 'SMS-received' from df
df.drop("SMS_received", axis=1, inplace=True)
#appedning the array to our dataframe with a new column 'SMS_recieved'
df['SMS_received'] = label

#Printing the dataframe
df

#Deleted the 'Date.diff' column
df.drop("Date.diff", axis=1)

#Scatter plot changed the x, y values into 'Age' and 'PatientID'
plt.scatter(df['Age'],df['PatientId'], c='blue', s=2)
plt.xlabel("Age")
plt.ylabel("PatientID")


# Value of k=2
k=2
X=np.array(list(zip(df['Age'], df['PatientId'])))
model=KMeans(n_clusters=k)
kmeans=model.fit(X)
labels=model.predict(X)
centroids=model.cluster_centers_


print(labels)
print(centroids)

#Plot the data and centriods on the scatter plot
#map the labels to colors
c=['blue','purple']
colors=[c[i] for i in labels]

plt.scatter(df['Age'], df['PatientId'], c=colors, s=2)
plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=100, c='black')


from sklearn.metrics import silhouette_score

model=KMeans(n_clusters=2)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=2): {silhouette_score(X,label)}')


model=KMeans(n_clusters=3)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=3): {silhouette_score(X,label)}')


model=KMeans(n_clusters=4)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=4): {silhouette_score(X,label)}')

# This dataset analyzes patient healthcare appointments, including age, ID number,
# SMS reminder status, and financial assistance.
# K-Means clustering was used to group similar patients after converting
# categorical columns into numeric form.
# Models with 2, 3, and 4 clusters were evaluated using a scoring metric.
# The results identify which grouping best separates patients and can
# support future improvements in scheduling and reminder systems.



```

```python
#Alexis Velasquez
#https://www.kaggle.com/datasets/avineshprabhakaran/loan-eligibility-predictionhttps://www.kaggle.com/datasets/avineshprabhakaran/loan-eligibility-prediction

#call the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# load the data
df=pd.read_csv("Loan Eligibility Prediction.csv")
df.head()

# info on the data
df.info()

# to know all the columns and rows
print(df.shape)

#To find how many na there is in the dataset
df.isnull().sum()

#remove the missing values with NaNs
df=df.dropna(subset=['Applicant_Income','Loan_Amount'])
print(df.shape)

#scatter plot
plt.scatter(df['Applicant_Income'],df['Loan_Amount'], c='red', s=2)
plt.xlabel("Applicant Income")
plt.ylabel("Loan Amount")

#For now, we'll try k value 2
k=2
X=np.array(list(zip(df['Applicant_Income'], df['Loan_Amount'])))
model=KMeans(n_clusters=k)
kmeans=model.fit(X)
labels=model.predict(X)
centroids=model.cluster_centers_

print(labels)
print(centroids)


#Plot the data and centroids on a scatter plot
#map the labels to colors

c=['m','g','b','r','y','c']
colors=[c[i] for i in labels]

plt.scatter(df['Applicant_Income'], df['Loan_Amount'], c=colors, s=2)
plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=100, c='black')

#Using the model that you have just trained,
#you can use it to predict what cluster for any data pt

cluster=model.predict([[5000,150]])[0]
print(c[cluster])


#To determine the silhouette coefficient/score

from sklearn.metrics import silhouette_score

model=KMeans(n_clusters=2)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=2): {silhouette_score(X,label)}')


model=KMeans(n_clusters=3)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=3): {silhouette_score(X,label)}')


model=KMeans(n_clusters=4)
model.fit(X)
label=model.predict(X)

print(f'silhouette Score (n=4): {silhouette_score(X,label)}')


k=4
X=np.array(list(zip(df['Applicant_Income'],df['Loan_Amount'])))

kmeans=KMeans(n_clusters=k)
kmeans=kmeans.fit(X)
labels=kmeans.predict(X)
centroids=kmeans.cluster_centers_

#map the labels to colors
c=['b','r','y','g','c','m']
colors=[c[i] for i in labels]

plt.scatter(df['Applicant_Income'],df['Loan_Amount'],c=colors,s=2)
plt.scatter(centroids[:,0],centroids[:,1],marker='*',s=100,c='black')

print(centroids)

#For this lab using K means clustering, I was able to group loan
#applicants based on their income and loan amount they are looking to
#borrow. Testing multiple values of k to see how I can divide the data
# The results showed that people with low income tend to borrow low amounts
#and people with higher income tend to borrow larger loans.Center points
#make it easy to see the differences. This analysis shows us how income
#level has an influence on loan amount.
```
